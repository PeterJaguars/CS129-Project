```python
import sklearn
import numpy as np
import pandas as pd
data = pd.read_csv('NFL-PBPDataCleaned.csv')
data
```

```python
data = data.drop(columns=['Unnamed: 0', 'game_id', 'desc', 'pass_length', 'pass_location', 'air_yards', 'run_location', 'run_gap', 'yards_gained', 'qb_scramble', 'qb_dropback', 'wp', 'vegas_wp', 'spread_line', 'total_line', 'roof', 'xpass'])
```

```python
data = data.dropna()
```

```python
y = data.pop('play_type')
```

```python
data
```

```python
length = len(data)
```

```python
length
```

```python
def posteam_type_to_bin(col):
    newcol = []
    for val in col:
        if val == 'home':
             newcol.append(0);
        else:
            newcol.append(1);
    
    return newcol
```

```python
def play_to_bin(col):
    newcol = []
    for val in col:
        if val == 'run':
             newcol.append(0);
        else:
            newcol.append(1);
    
    return newcol
```

```python
def roof_to_bin(col):
    newcol = []
    for val in col:
        if val == 'outdoors':
            newcol.append(0);
        else:
            newcol.append(1);
    
    return newcol
```

```python
def surface_to_bin(col):
    newcol = []
    for val in col:
        if val == 'grass':
            newcol.append(0);
        else:
            newcol.append(1);
    
    return newcol
```

```python
y = play_to_bin(y)
```

```python
data['posteam_type'] = posteam_type_to_bin(data['posteam_type'])
```

```python

data['surface'] = surface_to_bin(data['surface'])
```

```python
data
```

```python
for col in data:
    data[col] = (data[col] - pd.DataFrame.mean(data[col]))/(pd.DataFrame.max(data[col]) - pd.DataFrame.min(data[col]))
    
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
svmmodel = svm.SVC()
svmmodel.fit(X_train[0:1000], y_train[0:1000])
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train[0:10000], y_train[0:10000])
```

```python
svmmodel.score(X_train[0:1000], y_train[0:1000])
```

```python
svmmodel.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
y_train_pred = rfc.predict(X_train)
sklearn.metrics.accuracy_score(y_train, y_train_pred)
```

```python
y_test_pred = rfc.predict(X_test)
sklearn.metrics.accuracy_score(y_test, y_test_pred)
```
